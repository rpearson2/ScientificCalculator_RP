```python
import math

class Scientific_Calculator():
    def __init__(self, input1, input2):
        self.input1=input1
        self.input2=input2
    
    def addition(num1,num2):
        print("Result: ", num1+num2)
        
    def subtraction(num1,num2):
        print("Result: ",num1-num2)
    
    def multiplication(num1, num2):
        print("Result: ", num1*num2)
        
    def division(num1,num2):
        print("Result: ", num1/num2)
        
    def sqrt(num1):
        print("Result: ", math.sqrt(num1))
    
    def sin(num1):
        print("Result: ", math.sin(num1))
        
    def cos(num1):
        print("Result: ", math.cos(num1))
        
    def tan(num1):
        print("Result: ", math.tan(num1))
        
    def remainder(num1, num2):
        print("Result: ",math.remainder(num1, num2))
        
    def exp(num1):
        print("Result: ", math.exp(num1))
        
    def log10(num1):
        print("Result: ", math.log10(num1))
        
    def power(num1,num2):
        print("Result: ", math.pow(num1, num2))
        
    def acos(num1):
        print("Result: ", math.acos(num1))
        
    def asin(num1):
        print("Result: ", math.asin(num1))
    
    def atan(num1):
        print("Result: ", math.atan(num1))
        
    def acosh(num1):
        print("Result: ", math.acosh(num1))
        
    def asinh(num1):
        print("Result: ", math.asinh(num1))
        
    def atanh(num1):
        print("Result: ", math.atanh(num1))
        
    def factorial(num1):
        print("Result: ", math.factorial(num1))
        
while True:
    print("\nSelect a number: ")
    print("Enter '1' for addition")
    print("Enter '2' for subtraction")
    print("Enter '3' for multiplication")
    print("Enter '4' for division")
    print("Enter '5' for square root")
    print("Enter '6' for sin")
    print("Enter '7', for cos")
    print("Enter '8' for tan")
    print("Enter '9' for remainder")
    print("Enter '10' for exponential")
    print("Enter '11' for log10")
    print("Enter '12' for power")
    print("Enter '13' for acos")
    print("Enter '14' for asin")
    print("Enter '15' for atan")
    print("Enter '16' for acosh")
    print("Enter '17' for asinh")
    print("Enter '18' for atanh")
    print("Enter '19' for factorial")
    print("Enter '20' to quit")
    choice=int(input("Enter your choice (1-20): "))
    
    if choice==1:
        num1=int(input("Enter the first number: "))
        num2=int(input("Enter the second number: "))
        Scientific_Calculator.addition(num1,num2)
    elif choice==2:
        num1=int(input("Enter the first number: "))
        num2=int(input("Enter the second number: "))
        Scientific_Calculator.subtraction(num1,num2)
    elif choice==3:
        num1=int(input("Enter the first number: "))
        num2=int(input("Enter the second number: "))
        Scientific_Calculator.multiplication(num1,num2)
        
    elif choice==4:
        num1=int(input("Enter the first number: "))
        num2=int(input("Enter the second number: "))
        if num2==0:
            print("Infinity")
        else:
            Scientific_Calculator.division(num1,num2) 
    elif choice==5:
        num1=int(input("Enter the number: "))
        if num1<0:
            print("Invalid number, can't be less than 0.")
        else:
            Scientific_Calculator.sqrt(num1) 
            
    elif choice==6:
        num1=int(input("Enter the number: "))
        Scientific_Calculator.sin(num1) 
        
    elif choice==7:
        num1=int(input("Enter the number: "))
        Scientific_Calculator.cos(num1) 
    
    elif choice==8:
        num1=int(input("Enter the number: "))
        Scientific_Calculator.tan(num1) 
        
    elif choice==9:
        num1=int(input("Enter the first number: "))
        num2=int(input("Enter the second number: "))
        Scientific_Calculator.remainder(num1,num2)
    
    elif choice==10:
        num1=int(input("Enter the number: "))
        Scientific_Calculator.exp(num1) 
    
    elif choice==11:
        num1=int(input("Enter the number: "))
        Scientific_Calculator.log10(num1)
        
    elif choice==12:
        num1=int(input("Enter the first number: "))
        num2=int(input("Enter the second number: "))
        Scientific_Calculator.pow(num1,num2)
    
    elif choice==13:
        num1=int(input("Enter the number: "))
        Scientific_Calculator.acos(num1)
    
    elif choice==14:
        num1=int(input("Enter the number: "))
        Scientific_Calculator.asin(num1)
        
    elif choice==15:
        num1=int(input("Enter the number: "))
        Scientific_Calculator.atan(num1)
    
    elif choice==16:
        num1=int(input("Enter the number: "))
        Scientific_Calculator.acosh(num1)
        
    elif choice==17:
        num1=int(input("Enter the number: "))
        Scientific_Calculator.asinh(num1)
        
    elif choice==18:
        num1=int(input("Enter the number: "))
        Scientific_Calculator.atanh(num1)
     
    elif choice==19:
        num1=int(input("Enter the number: "))
        Scientific_Calculator.factorial(num1)

    elif choice==20:
        break
        
    else:
        print("Invalid number, please choose a number between (1-5)")
```
